```python
fecha_ultimo_conteo = "2019-11-03"

visits_df_1 = pd.read_csv('1/tabla_visitas_1.csv', index_col=0)
visits_df_1 = visits_df_1[visits_df_1['fecha'] > fecha_ultimo_conteo]
visits_df_2 = pd.read_csv('2/tabla_visitas_2.csv', index_col=0)
visits_df_2 = visits_df_2[visits_df_2['fecha'] > fecha_ultimo_conteo]
visits_df_3 = pd.read_csv('3/tabla_visitas_3.csv', index_col=0)
visits_df_3 = visits_df_3[visits_df_3['fecha'] > fecha_ultimo_conteo]
visits_df_4 = pd.read_csv('4/tabla_visitas_4.csv', index_col=0)
visits_df_4 = visits_df_4[visits_df_4['fecha'] > fecha_ultimo_conteo]

visits_dfs = [visits_df_1, visits_df_2, visits_df_3, visits_df_4]

for visits_df in visits_dfs:
    print(visits_df.info())
```

```python
print(len(visits_df_1.persona.unique()))
print(len(visits_df_2.persona.unique()))

print(min(visits_df_1.persona.unique()))
print(max(visits_df_1.persona.unique()))

print(min(visits_df_2.persona.unique()))
print(max(visits_df_2.persona.unique()))
```

```python
for visits_df in visits_dfs:
    
    persona = 1
    
    for person in np.sort(visits_df.persona.unique()):
        visits_df.loc[visits_df['persona'] == person, 'persona'] = persona
        persona += 1
        
visits_df_2.persona += max(visits_df_1.persona.unique())
visits_df_3.persona += max(visits_df_2.persona.unique())
visits_df_4.persona += max(visits_df_3.persona.unique())
```

```python
print(len(visits_df_1.persona.unique()))
print(len(visits_df_2.persona.unique()))

print(min(visits_df_1.persona.unique()))
print(max(visits_df_1.persona.unique()))

print(min(visits_df_2.persona.unique()))
print(max(visits_df_2.persona.unique()))
```

```python
print(len(visits_df_3.persona.unique()))
print(min(visits_df_3.persona.unique()))
print(max(visits_df_3.persona.unique()))
```

```python
visits_df = pd.concat(visits_dfs, ignore_index=True)
print(visits_df.persona.max())
visits_df.info()
```

```python
visits_df['grupo'] = 1

for i in visits_df.index[1:]:
    try:
        hora_ant = datetime.strptime(visits_df.loc[i-1, 'hora'], '%H:%M:%S.%f')
    except ValueError:
        hora_ant = datetime.strptime(visits_df.loc[i-1, 'hora'], '%H:%M:%S')
    try:
        hora = datetime.strptime(visits_df.loc[i, 'hora'], '%H:%M:%S.%f')
    except ValueError:
        hora = datetime.strptime(visits_df.loc[i, 'hora'], '%H:%M:%S')
                             
    misma_fecha = (visits_df.loc[i-1, 'fecha'] == visits_df.loc[i, 'fecha'])
    mismo_concesionario = (visits_df.loc[i-1, 'Concesionario'] == visits_df.loc[i, 'Concesionario'])
                             
    if ((hora - hora_ant).seconds < 15) & misma_fecha & mismo_concesionario:
        visits_df.loc[i, 'grupo'] = visits_df.loc[i-1, 'grupo']
    else: 
        visits_df.loc[i, 'grupo'] = visits_df.loc[i-1, 'grupo'] + 1
```
